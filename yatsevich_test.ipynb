### Напиши функцию, которая удалит дубликаты в списке. Список не отсортирован. Необходимо сохранить порядок сортировки оригинального списка.


```python
'Асимптотическая сложность порядка O(N), где N -- количество элементов'

def delDuplicates(a):
    new = []
    s = set()
    for elem in a:
        if elem in s:
            pass
        else:
            new.append(elem)
            s.add(elem)
    return new
```

```python
a = [1, 3, 2, 1, 5, 3, 5, 1, 4]
print(delDuplicates(a))
```

### Подкинули монету N раз. Кол-во случаев, когда выпал орёл, на 10% больше, чем кол-во случаев, когда выпала решка. При каком N мы можем сказать, что монета «нечестная» (орёл и решка выпадают с разной вероятностью)?


###  Задание 3
Подкинули монету N раз. Кол-во случаев, когда выпал орёл, на 10% больше, чем кол-во случаев, когда выпала решка. При каком N мы можем сказать, что монета «нечестная» (орёл и решка выпадают с разной вероятностью)?


Формулируем две гипотезы.

Hулевая гипотеза: вероятность выпадения орла и решки одинакова (p = 0.5);

альтернатива: вероятность не одинакова.

Известно, что по результатам испытаний, количество случаев, когда выпадает орел, на 10% больше.

n - количество испытаний всего, k - количество случаев выпадения орла.

pVal - вероятность того, что нулевая гипотеза не верна.

Уровень значимости, с которым можно отвергнуть нулевую гипотезу возьму равным 0.05.

В цикле расчитывается p-value до тех пор, пока оно не станет меньше уровня значимости 0.05, на каждом шаге прибавляю количество проведенных экспериментов.

```python
from scipy import stats
```

```python
n = 10#начнем с десяти испытаний
pVal = 1.
while (pVal>=0.05):
    n += 1# n - количество испытаний всего
    k = round(n*11/20)# k - количество случаев выпадения орла
    pVal = stats.binom_test(k, n, 0.5, alternative='greater')# pVal - вероятность того, что нулевая гипотеза не верна.
    print('n = ', n, 'k = ', k, 'p-value = ', round(pVal,3), sep='\t')
```

Получено количество испытаний n = 281, при таком количестве испытаний можно сказать, что монета "нечестная" с уровнем значимости 0.05.
